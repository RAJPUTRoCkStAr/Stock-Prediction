```python
import folium
from geopy.geocoders import Nominatim
import yfinance as yf

def get_company_info(ticker_symbol):
    company = yf.Ticker(ticker_symbol)
    company_info = {
        'city': company.info.get('city', 'N/A'),
        'state': company.info.get('state', 'N/A'),
        'country': company.info.get('country', 'N/A'),
        'zip': company.info.get('zip', 'N/A'),
    }
    return company_info

def get_coordinates(address):
    geolocator = Nominatim(user_agent="my_geocoder")
    location = geolocator.geocode(address)
    if location:
        return location.latitude, location.longitude
    else:
        return None, None

ticker_symbol = 'ATS'
company_info = get_company_info(ticker_symbol)

address = f"{company_info['city']}, {company_info['state']}, {company_info['country']}"
latitude, longitude = get_coordinates(address)

if latitude is not None and longitude is not None:
    map = folium.Map(location=[latitude, longitude], zoom_start=10)
    folium.Marker(
        [latitude, longitude],
        popup=f'{ticker_symbol} Headquarters',
        tooltip='Click for more info'
    ).add_to(map)
    display(map)
else:
    print("Coordinates not found for the company's location.")
```

```python
import folium
from geopy.geocoders import Nominatim


def get_company_info(ticker_symbol):
    company = yf.Ticker(ticker_symbol)
    company_info = {
        'city': company.info.get('city', 'N/A'),
        'state': company.info.get('state', 'N/A'),
        'country': company.info.get('country', 'N/A'),
        'zip': company.info.get('zip', 'N/A'),
    }
    return company_info

def get_coordinates(address):
    geolocator = Nominatim(user_agent="my_geocoder")
    location = geolocator.geocode(address)
    if location:
        return location.latitude, location.longitude
    else:
        return None, None

ticker_symbol = 'ATS'
company_info = get_company_info(ticker_symbol)

address = f"{company_info['city']}, {company_info['state']}, {company_info['country']}"
latitude, longitude = get_coordinates(address)

if latitude is not None and longitude is not None:
    map = folium.Map(location=[latitude, longitude], zoom_start=10)
    folium.Marker(
    [latitude, longitude],
    popup=f'{ticker_symbol} Headquarters',
    tooltip='Click for more info',
    icon=folium.Icon(color='blue')
    ).add_to(map)
    folium.Marker(
    [latitude, longitude],
    popup=f'{company_info["country"]}',
    tooltip=f'{company_info["country"]}',
    icon=folium.Icon(color='red', icon='info-sign')
    ).add_to(map)
    display(map)
else:
    print("Coordinates not found for the company's location.")
```
