```python
import yfinance as yf
# Specify the stock ticker symbol and the time period
stock_symbol = "ATS"
start_date = "2022-01-01"
end_date = "2022-12-31"
# Download historical data
stock_data = yf.download(stock_symbol, start=start_date, end=end_date)
# Print the first few rows of the data
print(stock_data.head())
```

```python
ticker = yf.Ticker('ATS')
for key, value in ticker.info.items():
    print(f'{key}: {value}')
```

```python
import matplotlib.pyplot as plt
stock_data['Close'].plot(figsize=(10, 6), title=f'{stock_symbol} Stock Price')
plt.xlabel('Date')
plt.ylabel('Closing Price (USD)')
plt.show()
```

```python
import yfinance as yf
def get_company_info(ticker_symbol):
    # Create a Ticker object for the company
    company = yf.Ticker(ticker_symbol)
    # Get various details about the company
    company_info = {
        'Name': company.info.get('longName', 'N/A'),
        'Symbol': ticker_symbol,
        'Industry': company.info.get('industry', 'N/A'),
        'Sector': company.info.get('sector', 'N/A'),
        'Market Cap': company.info.get('marketCap', 'N/A'),
        'Previous Close': company.info.get('regularMarketPreviousClose', 'N/A'),
        'Open': company.info.get('regularMarketOpen', 'N/A'),
        'Dividend Yield': company.info.get('dividendYield', 'N/A'),
        'Currency': company.info.get('currency', 'N/A'),
        'P/E Ratio': company.info.get('trailingPE', 'N/A'),
        'Volume': company.info.get('volume', 'N/A'),
        # Add more details as needed
    }
    return company_info
# Example: Get information about Apple Inc. (AAPL)
ticker_symbol = 'ATS'
apple_info = get_company_info(ticker_symbol)
# Print the information
for key, value in apple_info.items():
    print(f'{key}: {value}')
```
